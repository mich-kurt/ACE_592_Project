```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
## Creating a column for sentiment scores
sid = SentimentIntensityAnalyzer()
df['sentiment'] = [sid.polarity_scores(x)['compound'] for x in df.Comment]
```

```python
data['sentiment'] = df[['sentiment']]
data['Drug'] = df['Drug'].to_list()
```

```python
data['Drug'].unique()
# create separate datas for each drug
drug_1 = data[data['Drug'] == 'Ozempic']
drug_2 = data[data['Drug'] == 'Sitagliptin']
drug_3 = data[data['Drug'] == 'Mounjaro']
drug_4 = data[data['Drug'] == 'Rybelsus']
drug_5 = data[data['Drug'] == 'Victoza']
drug_6 = data[data['Drug'] == 'Jardiance']

drug_1.drop(['Drug'], axis = 1, inplace = True)
drug_2.drop(['Drug'], axis = 1, inplace = True)
drug_3.drop(['Drug'], axis = 1, inplace = True)
drug_4.drop(['Drug'], axis = 1, inplace = True)
drug_5.drop(['Drug'], axis = 1, inplace = True)
drug_6.drop(['Drug'], axis = 1, inplace = True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import ElasticNet, ElasticNetCV
from sklearn.linear_model import LassoCV
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

def get_coefs(alpha,features,target):
    X = StandardScaler().fit_transform(data[features])
    y = StandardScaler().fit_transform(data[target])
    
    lasso_model =Lasso(alpha=alpha)
    lasso_model = lasso_model.fit(y=y,X=X)
    
    coefs = lasso_model.coef_
    
    return coefs
```
